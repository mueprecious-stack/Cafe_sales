ANALYSIS AND VISUALIZATION OF CLEANED CAFE SALES DATASET

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Cleaned Dataset

```python
df = pd.read_csv('cleaned_cafe_sales.csv')
df.head()
```

```python
df.describe()
```

Performance of Products

```python
#monthly sales
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
monthly_sales = df.groupby('Month')['Total Spent'].sum()
sns.lineplot(monthly_sales, color='black', linewidth=2, marker='o', markersize=9)
plt.title('Monthly Sales Performance')

#Performance by Total Spent
items_by_sales = df.groupby('Item')['Total Spent'].sum()
plt.subplot(1,2,2)
sns.barplot(items_by_sales, color='teal')
plt.title('Items by Sales')


plt.figure(figsize=(20,5))

#correlation between numerical variables
plt.subplot(1,3,1)
correlation = df[['Quantity','Price Per Unit','Total Spent']]
sns.heatmap(correlation.corr(), annot=True)
plt.title('Correlation')
plt.show

#Performance based on quantity of items ordered
plt.subplot(1,3,2)
quantity_ordered = df.groupby('Item')['Quantity'].sum()
plt.pie(quantity_ordered.values,labels=quantity_ordered.index, autopct='%1.1f%%')
plt.title('Most Purchased Items')

#Customers' location preference
plt.subplot(1,3,3)
location_preference = df.groupby('Location')['Transaction ID'].count()
plt.pie(location_preference.values,labels= location_preference.index, autopct = '%1.1f%%')
plt.title('Most Preferred Location')
plt.show


```

Business insights

* Sales peak in June and decline in February and November.
* Higher-priced items tend to generate higher sales, as shown by the strong positive correlation (0.65) between 'Price Per Unit' and 'Total Spent'.
* Item price appears to have little influence on customer purchasing behaviour as indicated by the very weak positive correlation (0.0073) between 'Price Per Unit' and 'Quantity' ordered.
* Sales revenue increase as the number of items ordered increases, indicating a direct positive correlation(0.70) between 'Quantity' and 'Total Spent'
* Coffee contributes the most to overall sales at 12.9%, while Cookie contributes the least at 11.9%
* Customers prefer in-store orders to Takeaway orders


Recommendations
* Allocate more inventory and staff during the June peak, and increase marketing efforts during declines in February and November
* Since customer purchasing behaviour does not appear to be significantly affected by item prices, introducing premium packages could be a viable strategy to increase overall sales and profit.
* Given the high demand for coffee, increasing its quantity can maximize sales.
* Customers Preference for in-store orders suggests that openning more branches or expanding the cafe could drive growth
